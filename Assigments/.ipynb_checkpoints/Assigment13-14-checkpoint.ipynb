#### For this bivariate analysis, at least perform the following tasks:

```python
import pandas as pd
import seaborn as sns
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
PremStats = pd.read_csv("../data/EPL_20_21.csv")
```

```python
contingencyTable1 = penguins.groupby(['species','sex']).size().unstack('species')
contingencyTable1
```

```python
contingencyTable2 = penguins.groupby(['island','sex']).size().unstack('island')
contingencyTable2
```

Ik verwacht wel een correlatie is aangezien de vederling vaak 50/50 is per sex

```python
contingencyTable1.plot(kind= 'bar')
```

```python
contingencyTable2.plot(kind= 'bar')
```

```python

chi2_contingency(contingencyTable1)
```

```python

chi2_contingency(contingencyTable2)
```
